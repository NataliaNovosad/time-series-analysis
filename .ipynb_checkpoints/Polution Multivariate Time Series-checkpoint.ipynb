```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss
import re
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data = pd.read_csv("time-series-datasets/AirQualityUCI/AirQualityUCI.csv", sep = ";")
data = data.drop(["Unnamed: 15", "Unnamed: 16"], axis = 1)
data = data.dropna(how = "any")
data = data.applymap(lambda x: str(x).replace(",", "."))
data["Datetime"] = data["Date"] + "/" + data["Time"]
```

```python
data.info()
```

```python
datetime = pd.to_datetime(data["Datetime"], format = "%d/%m/%Y/%M.%S.%f")
data.index = datetime
data = data.drop(["Date", "Time", "Datetime"], axis = 1)
```

```python
data.head()
```

## Column meaning

1) True hourly averaged concentration CO in mg/m3  CO(GT); 

2) PT08.S1 (tin oxide) hourly averaged sensor response  PT08.S1(CO); 

3) True hourly averaged Benzene concentration in microg/m3  C6H6(GT); 

4) PT08.S2 (titania) hourly averaged sensor response  PT08.S2(NMHC); 

5) True hourly averaged NOx concentration in ppb  NOx(GT); 

6) PT08.S3 (tungsten oxide) hourly averaged sensor response  PT08.S3(NOx); 

7) True hourly averaged NO2 concentration in microg/m3  NO2(GT); 

8) PT08.S4 (tungsten oxide) hourly averaged sensor response  PT08.S4(NO2); 

9) PT08.S5 (indium oxide) hourly averaged sensor response  PT08.S5(O3); 

10) Temperature in °C  T; 

11) Relative Humidity (%)  RH; 

12) Absolute Humidity  AH.

```python
for cl in data.columns:
    for i in range(data.shape[0]):
        if data[cl][i] < 0:
            data[cl][i] = data[cl][i-1]
```

```python
data = data.astype(float)

for cl in data.columns:
    plt.figure(figsize = (20,5))
    data[cl].plot(title = cl)
    plt.show()
```

```python
import seaborn as sns
corr_data = data.corr()
sns.heatmap(corr_data)
```
